```python
def numerical_dis_plot(df, features, segment_feature=None):
    fig, axs = plt.subplots(nrows=len(features), ncols=1, figsize=(8, 30))
    
    for i, feature in enumerate(features):
        if segment_feature:
            sns.boxplot(x=feature, y=segment_feature, data=df, ax=axs[i], showfliers=True)
        else:
            sns.boxplot(x=feature, data=df, ax=axs[i], showfliers=True)
    plt.tight_layout()
    plt.show()
    
def categorical_plot(df, features, segment_feature=None):
    fig, axs = plt.subplots(nrows=len(features), ncols=1, figsize=(12,20))
    
    for i, feature in enumerate(features):
        if segment_feature:
            sns.countplot(data=df, x=segment_feature, hue=feature, ax=axs[i])
        else:
            sns.countplot(data=df, hue=feature, ax=axs[i])
    plt.tight_layout()
    plt.show()
```

Terlihat dengan jelas pada fitur `PCA_CareerFinance`, `PCA_YearsExperience`, dan `PCA_PerformanceSalaryHike` memiliki persebaran yang tidak merata (_skewed_). Untuk mengatasi permasalahan ini, akan menggunakan `PowerTransformer` untuk me-normalisasi persebaran datanya bersamaan dengan fitur-fitur numerik lainnya.

```python
main_df.info()
```

```python
new_numeric_features = ["Age","DistanceFromHome","EnvironmentSatisfaction",
                        "JobSatisfaction","NumCompaniesWorked",
                        "RelationshipSatisfaction","StockOptionLevel",
                        "PCA_CareerFinance","PCA_YearsExperience",
                        "PCA_PerformanceSalaryHike",
                       ]
```

```python
from sklearn.preprocessing import PowerTransformer

def power_transforms(df, features):
    transformed_df = df.copy()
    transformed_feature_list = []
    for feature in features:
        
        # Initialize the PowerTransformer
        transformer = PowerTransformer(standardize=True)
        
        # Transform the value of the feature into numpy array
        y = np.asanyarray(transformed_df[feature])
        
        # Transform its numpy array to have a single column to be fitted in the transformer
        y = y.reshape(-1,1)
        
        # Transform the data distribution into gaussian-like distribution
        transformer.fit(y)
        
        # Save the value into a new column
        transformed_df[f"transformed_{feature}"] = transformer.transform(y)
        
        # Drop current feature
        transformed_df.drop(feature, axis=1, inplace=True)
        
        # Save the fitted transformer
        joblib.dump(transformer, f"Transformed_{feature}")
        
        transformed_feature_list.append(f"transformed_{feature}")
    return transformed_df, transformed_feature_list
```

```python
transformed_df, transformed_feature_list = power_transforms(main_df, new_numeric_features)
transformed_df.head()
```

```python
numerical_dis_plot(transformed_df, transformed_feature_list)
```

## Modeling

Dua kolom pertama dari _DataFrame_ ini adalah `EmployeeId` dan `Attrition`. Untuk kolom `EmployeeId`, tidak akan digunakan dalam proses pelatihan model. Untuk kolom `Attrition`, akan digunakan sebagai target dari model ini.

```python
without_null_df = transformed_df[~(transformed_df["Attrition"].isnull())].copy()
with_null_df = transformed_df[transformed_df["Attrition"].isnull()].copy()

print("Length of dataframe without null:", without_null_df.shape[0])
print("Length of dataframe with null:", with_null_df.shape[0])
```

```python
without_null_df
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression

X, y = without_null_df.iloc[:,2:], without_null_df["Attrition"]
```

### Logistic Regression

```python
kfold = KFold(n_splits=4, shuffle=True, random_state=42)

LRmodel = LogisticRegression()

scores = cross_val_score(LRmodel, X, y, cv=kfold, scoring="accuracy")
LRmodel_mean_accuracy = np.mean(scores)

print("Cross-Validation score:", scores)
print("Mean Accuracy:", LRmodel_mean_accuracy)
```

```python
joblib.dump(LRmodel, "LR_Attrition Predictor model.joblib")
```

### Deep Learning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
leaky_relu = LeakyReLU(alpha=0.2)

DLmodel = Sequential([
    Dense(128, activation=leaky_relu),
    Dense(32, activation=leaky_relu),
    Dense(1, activation="sigmoid")
])
```

```python
DLmodel.compile(loss="binary_crossentropy",
              optimizer=Adam(learning_rate=2.5e-3),
              metrics=["acc"])
```

```python
fold_scores = []
for train_index, val_index in kfold.split(X):
    
    X_train_fold, X_val_fold = X.iloc[train_index], X.iloc[val_index]
    y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[val_index]

    # Train the model on the current fold
    DLmodel.fit(X_train_fold, y_train_fold, epochs=25, validation_data=(X_val_fold, y_val_fold))

    # Evaluate the model on the validation data for this fold
    fold_loss, fold_accuracy = DLmodel.evaluate(X_val_fold, y_val_fold)
    fold_scores.append(fold_accuracy)
```

```python
DLmodel.save("DL_Attrition Predictor model.h5")
```

```python
DLmodel_mean_accuracy = np.mean(fold_scores)
print("Mean Accuracy:", DLmodel_mean_accuracy)
```

### Random Forest Classifier

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25)

RFmodel = RandomForestClassifier()

param_grid = {
    "n_estimators":[100,200,300],
    "max_depth":[None,50,150],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}

strat_kfold = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)
```

```python
grid_search = GridSearchCV(RFmodel, param_grid, cv=kfold, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
RFmodelbest = grid_search.best_estimator_
RFmodelpredictions = RFmodelbest.predict(X_val)

RFmodel_accuracy = accuracy_score(y_val, RFmodelpredictions)
print("Random Forest Classifier's Accuracy:", RFmodel_accuracy)
```

```python
joblib.dump(RFmodelbest, "RF_Attrition Predictor model.joblib")
```

Setelah dilakukan imputasi pada fitur `Attrition`, saatnya untuk melakukan _Clustering_.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = {}
silhouette = {}

for k in range(2,10):
    model = KMeans(n_clusters=k, random_state=123)
    
    # Exclude the `EmployeeId` feature (first column)
    model.fit_predict(imputed_df.iloc[:,1:])
    
    inertia[k] = model.inertia_
    silhouette[k] = silhouette_score(imputed_df.iloc[:,1:], model.labels_)

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(11,4))
axs[0].plot(inertia.keys(), inertia.values(), marker="o", lw=3)
axs[0].set_xlabel("Number of Clusters", fontsize=12)
axs[0].set_ylabel("Inertia score", fontsize=12)
axs[0].set_title("The Elbow Method", fontsize=15)

axs[1].plot(silhouette.keys(), silhouette.values(), marker="*", lw=3)
axs[1].set_xlabel("Number of Clusters", fontsize=12)
axs[1].set_ylabel("Silhouette score", fontsize=12)
axs[1].set_title("The Silhouette Method", fontsize=15)

plt.show()
```

Pada metode **Elbow**, perubahan nilai mulai melandai dari titik 4. Artinya, jumlah klaster yang optimal berdasarkan metode ini adalah 4 klaster.

Sedangkan untuk metode **Silhouette coefficient**, nilai siluet tertinggi (selain 2) ada pada titik 3 kemudian 4.

Berdasarkan kedua metode tersebut, akan diambil kesimpulan yang beririsan, yaitu jumlah klaster yang optimal adalah 4 klaster.

```python
model = KMeans(n_clusters=4, random_state=123)
model.fit(imputed_df.iloc[:,1:])
joblib.dump(model, "kmeans_clustering_model.joblib")
```

Setelah selesai melakukan klasterisasi berdasarkan jumlah klaster yang optimal, lakukan prediksi nilai klaster dan tambahkan pada _dataframe_ `df`.

```python
clusters = model.predict(imputed_df.iloc[:,1:])
df["EmployeeSegment"] = clusters.astype(str)
df
```
